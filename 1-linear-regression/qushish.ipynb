```python
data_qushish_jadvali = [ [a, b, a+b] for a in range(1, 11) for b in range(1, 11)]
print(data_qushish_jadvali)
```

### Datani numpyga o'tkazamiz

```python
import numpy as np

data = np.array(data_qushish_jadvali)
print(data.shape)
```

### Datani bo'lamiz xususiyatlar uchun va label uchun.

```python
X = data[:, :2]
y = data[:, 2]

print(X.shape)
print(y.shape)
```

### O'rgatish jarayoniga tayyormiz

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X, y)
```

### Bashorat qilish yangi data uchun
Masalan:

1) 5+6 = ?

2) 10+11 = ?

```python
yangi_data = np.array([[5, 6]])
natija = model.predict(yangi_data)
print(natija)
```

#Hammasi birgalikda

```python
import numpy as np
from sklearn.linear_model import LinearRegression


# data
data_qushish_jadvali = [ [a, b, a+b] for a in range(1, 11) for b in range(1, 11)]

# numpy
data = np.array(data_qushish_jadvali)
print(data.shape)

X = data[:, :2]
y = data[:, 2]

print(X.shape)
print(y.shape)

model = LinearRegression()
model.fit(X, y)

yangi_data = np.array([[5, 6]])
natija = model.predict(yangi_data)
print(natija)
```
